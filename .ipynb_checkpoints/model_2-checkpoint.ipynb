```python
import pandas as pd
import numpy as np

updated_vegetation_data = pd.read_csv("data/vegetation_data.csv")

used_columns = ['Location','Year', 'MM_CA', 'MM_CEC', 'MM_K', 'MM_MG', 'MM_OM',
       'MM_P1', 'MM_PH', 'MM_CLAY', 'MM_SAND', 'MM_SILT', 'Plot', 'Yield', 'X', 'Population', 'Cl', 'Clre', 'Datt1',
       'Datt4', 'Datt6', 'DDI', 'DPI', 'GNDVI', 'MCARI', 'MCARI3', 'mSR',
       'mSR2', 'MTCI', 'MTVI1', 'ND1', 'ND2', 'NDchl', 'NDRE', 'NDVI1',
       'NDVI2', 'NDVI3', 'NPCL', 'NPQI', 'OSAVI', 'PBI', 'PPR', 'PRI',
       'PSNDb1', 'PSNDc1', 'PSNDc2', 'PSRI', 'PSSRc1', 'PSSRc2', 'PVR', 'PWI',
       'RDVI', 'RVSI', 'SAVI', 'SIPI', 'SR1', 'SR2', 'SR3', 'SR4', 'DSWI-4',
       'SRPI', 'TCARI', 'TCI', 'TVI', 'WBI', 'Latitude', 'Longitude']

updated_vegetation_data = updated_vegetation_data[used_columns]
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial import distance_matrix

# Compute pairwise distances based on latitude and longitude
distances = distance_matrix(updated_vegetation_data[['Latitude', 'Longitude']].values, 
                            updated_vegetation_data[['Latitude', 'Longitude']].values)

# Set a threshold based on the 5th percentile of distances (excluding zero)
threshold = np.percentile(distances[distances > 0], 3)

# Create edges based on the threshold
spatial_edges = np.argwhere(distances < threshold)

# Remove self-loops (i.e., where source and target nodes are the same)
spatial_edges = spatial_edges[spatial_edges[:, 0] != spatial_edges[:, 1]]

spatial_edges[:10], len(spatial_edges)  # Displaying first 10 edges and total number of spatial edges
```

```python
from itertools import combinations

# Modified function to generate genotype-based edges
def create_genotype_edges_all_pairs(data, column):
    """
    Generate edges based on a specific column (e.g., genotype or population) such that
    all nodes with the same value in the column have an edge between them.
    
    Args:
    - data (DataFrame): The dataset containing the column.
    - column (str): Name of the column to base the edges on.
    
    Returns:
    - np.array: Array of edges.
    """
    edges = []
    
    # Extract unique values from the specified column
    unique_values = data[column].unique()
    
    for value in unique_values:
        # Extract nodes associated with the current unique value
        nodes = data[data[column] == value].index.values
        
        # Generate all possible edges among these nodes
        all_edges = list(combinations(nodes, 2))
        
        # Append the edges to the list
        edges.extend(all_edges)
    
    return np.array(edges)

# Generate genotype-based edges using the Population column
genotype_edges_all_pairs = create_genotype_edges_all_pairs(updated_vegetation_data, 'Population')

# Displaying first 10 edges and total number of genotype edges
genotype_edges_all_pairs[:10], len(genotype_edges_all_pairs)
```

```python
import torch
from torch_geometric.data import Data
import pandas as pd

def combine_edges(spatial, genotype):
    """Combine spatial and genotype edges."""
    return np.vstack([spatial, genotype])

def to_tensor(array, dtype):
    """Convert a NumPy array to a PyTorch tensor."""
    return torch.tensor(array, dtype=dtype)

def extract_node_features(data, feature_cols, categorical_cols=None):
    """Extract node features from a DataFrame."""
    if categorical_cols:
        data = pd.get_dummies(data, columns=categorical_cols)
        # Update feature columns to include new one-hot encoded columns and exclude the original categorical columns
        feature_cols = [col for col in data.columns if col not in categorical_cols]
    return data[feature_cols].values


def create_graph_data(node_features, all_edges, target_label):
    """Create a PyTorch Geometric Data object."""
    x = to_tensor(node_features, dtype=torch.float)
    edge_index = to_tensor(all_edges, dtype=torch.long).t().contiguous()
    y = to_tensor(target_label, dtype=torch.float).unsqueeze(1)
    return Data(x=x, edge_index=edge_index, y=y)

# Combine spatial and genotype edges
all_edges = combine_edges(spatial_edges, genotype_edges_all_pairs)
# Assuming `updated_vegetation_data` is your original DataFrame
updated_vegetation_data.fillna(updated_vegetation_data.mean(), inplace=True)

# Extract node features
numerical_cols = updated_vegetation_data.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_cols = ['Population', 'Location']
feature_cols = numerical_cols + categorical_cols

node_features = extract_node_features(updated_vegetation_data, feature_cols, categorical_cols)

# Extract target label
target_label = updated_vegetation_data['Yield'].values

# Create PyTorch Geometric Data object
graph_data = create_graph_data(node_features, all_edges, target_label)

# Display graph_data (This line is for demonstration and should be included only if you want to see the Data object)
print(graph_data)
```

```python
# Assume graph_data.x contains all features including 'Yield' at index 0
features_to_scale = graph_data.x[:, 1:]  # Exclude the 'Yield' column assumed to be at index 0

# Calculate mean and std
mean = torch.mean(features_to_scale, dim=0)
std = torch.std(features_to_scale, dim=0)

# Perform standard scaling
scaled_features = (features_to_scale - mean) / (std + 1e-6)  # Adding a small constant to avoid division by zero

# Update the PyTorch Geometric Data object
graph_data.x[:, 1:] = scaled_features
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import SAGEConv

class EnhancedGraphSAGE(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super(EnhancedGraphSAGE, self).__init__()
        
        # Initial GraphSAGE layer
        self.conv1 = SAGEConv(in_channels, hidden_channels)
        
        # Additional GraphSAGE layers
        self.conv2 = SAGEConv(hidden_channels, hidden_channels)
        self.conv3 = SAGEConv(hidden_channels, hidden_channels)
        
        # Final GraphSAGE layer for prediction
        self.conv4 = SAGEConv(hidden_channels, out_channels)
        
        # Batch normalization layers for the hidden layers
        self.bn1 = nn.BatchNorm1d(hidden_channels)
        self.bn2 = nn.BatchNorm1d(hidden_channels)
        self.bn3 = nn.BatchNorm1d(hidden_channels)

    def forward(self, x, edge_index):
        # Initial layer
        x0 = F.relu(self.conv1(x, edge_index))
        x0 = self.bn1(x0)
        x0 = F.dropout(x0, p=0.5, training=self.training)
        
        # Second layer with skip connection
        x1 = F.relu(self.conv2(x0, edge_index) + x0)
        x1 = self.bn2(x1)
        x1 = F.dropout(x1, p=0.5, training=self.training)
        
        # Third layer with skip connection
        x2 = F.relu(self.conv3(x1, edge_index) + x1)
        x2 = self.bn3(x2)
        x2 = F.dropout(x2, p=0.5, training=self.training)
        
        # Final layer to produce output
        out = self.conv4(x2, edge_index)
        
        return out

# Define the dimensions
in_channels = graph_data.x.size(1)  # Number of input features (node features)
hidden_channels = 64  # Size of the hidden layer
out_channels = 1  # We're performing regression, so we have one output

# Instantiate the enhanced model
model = EnhancedGraphSAGE(in_channels, hidden_channels, out_channels)
```

```python
# Assuming the total number of nodes is num_nodes
num_nodes = graph_data.num_nodes

# Create masks for training, validation, and test sets
train_mask = torch.zeros(num_nodes, dtype=torch.bool)
val_mask = torch.zeros(num_nodes, dtype=torch.bool)
test_mask = torch.zeros(num_nodes, dtype=torch.bool)

# Assuming 70% training, 15% validation, 15% test
train_mask[:int(num_nodes * 0.7)] = 1
val_mask[int(num_nodes * 0.7):int(num_nodes * 0.85)] = 1
test_mask[int(num_nodes * 0.85):] = 1
```

```python
def train_one_epoch(model, optimizer, data, train_mask):
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)
    loss = F.mse_loss(out[train_mask], data.y[train_mask])
    loss.backward()
    optimizer.step()
    return loss.item()
```

```python
def evaluate_model(model, data, mask):
    model.eval()
    with torch.no_grad():
        pred = model(data.x, data.edge_index)
        loss = F.mse_loss(pred[mask], data.y[mask])
    return loss.item()
```

```python
# Initialize the model and optimizer
model = EnhancedGraphSAGE(graph_data.num_features, 64, 1)  # Assuming 64 hidden channels
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)  # Assuming a learning rate of 0.01

# Training Loop
for epoch in range(100):  # 100 epochs
    # Train model
    train_loss = train_one_epoch(model, optimizer, graph_data, train_mask)
    
    # Validate model
    val_loss = evaluate_model(model, graph_data, val_mask)
    
    # Print epoch info
    print(f"Epoch: {epoch+1}, Train Loss: {train_loss}, Validation Loss: {val_loss}")

# Test model
test_loss = evaluate_model(model, graph_data, test_mask)
print(f"Test Loss: {test_loss}")
```

```python
# 1. Model Architecture Explanation:
# The Enhanced GraphSAGE model is designed to capture both local and broader neighborhood information of nodes for prediction tasks.

# SAGEConv Layers: These are the primary layers in our GraphSAGE model. They operate by aggregating information from neighboring nodes, transforming this aggregated data, and then combining it with the node's own features.

# Batch Normalization: After each SAGEConv layer, we apply batch normalization. This helps in stabilizing and accelerating the training of deep networks. It normalizes the output of each layer to have a mean of zero and a variance of one.

# ReLU Activation: After normalization, we apply the Rectified Linear Unit (ReLU) activation function. It introduces non-linearity to the model, allowing it to capture complex patterns.

# Dropout: It's a regularization technique where randomly selected neurons are ignored during training. This helps in preventing overfitting.

# Skip Connections: These are shortcuts that bypass one or more layers. They help in combating the vanishing gradient problem in deeper networks and enhance the flow of information.

# In the architecture:

# The first layer captures immediate neighborhood information.
# Subsequent layers capture broader neighborhood information, with each layer effectively doubling the "reach" of neighborhood information considered.
# Skip connections help in retaining the information from previous layers, allowing the model to use both shallow and deep node representations.
```
